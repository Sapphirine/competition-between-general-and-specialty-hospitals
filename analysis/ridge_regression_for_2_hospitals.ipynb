```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
```

```python
general = pd.read_csv('general_hip.csv')
joint = pd.read_csv('joint_hip.csv')

label_temp = list(general['Total Charges'])
label = []
for i in label_temp:
    label.append(float(i[1:]))
num_variable = len(label)
label = np.array(label)/general['Length of Stay']
del general['Total Charges']
del general['Length of Stay']
l = list(general.columns.values)
variable = []
for i in l[1:]:
    variable.append(list(general[i]))

dic = {'18 to 29':1, '30 to 49':2, '50 to 69':3, '70 or Older':4}
temp = []
class_name = ['age']
for age in variable[0]:
    temp.append(dic[age])
feature = np.array(temp).reshape((num_variable,1))
for (N,i) in enumerate(variable[1:]):
    if N == 6:
        feature = np.hstack((feature,np.array(i).reshape((num_variable,1))))
        class_name.append('APR Severity of Illness Code')
    else:
        temp = mlb.fit_transform(zip(i))
        feature = np.hstack((feature,temp))
        class_name += list(mlb.classes_)

order_feature = range(len(feature))
np.random.shuffle(order_feature)
train = feature[order_feature[:1000],:]
test = feature[order_feature[1000:],:]
train_label = label[order_feature[:1000]]
test_label = label[order_feature[1000:]]

clf = RandomForestRegressor(n_estimators = 300,max_features = 0.1)
clf = clf.fit(train, train_label)
p = clf.predict(test)

print('random forest: ' + str(np.sum(abs(p - test_label))/np.sum(test_label)))
print('')
imp = clf.feature_importances_
for (N,i) in enumerate(imp):
    print(str(class_name[N]) + ': ' + str(i))
    
clf = Ridge(alpha = 88)
clf = clf.fit(train, train_label)
p = clf.predict(test)
print('')
print('redge regression: '+str(np.sum(abs(p - test_label))/np.sum(test_label)))
print('')
imp = clf.coef_
for (N,i) in enumerate(imp):
    print(str(class_name[N]) + ': ' + str(i))
g_Series = pd.Series(imp,index = class_name)
```

```python
general = pd.read_csv('general_hip.csv')
joint = pd.read_csv('joint_hip.csv')

label_temp = list(joint['Total Charges'])
label = []
for i in label_temp:
    label.append(float(i[1:]))
num_variable = len(label)
label = np.array(label)/joint['Length of Stay']
del joint['Total Charges']
del joint['Length of Stay']
l = list(joint.columns.values)
variable = []
for i in l[1:]:
    variable.append(list(joint[i]))

dic = {'0 to 17':0,'18 to 29':1, '30 to 49':2, '50 to 69':3, '70 or Older':4}
temp = []
class_name = ['age']
for age in variable[0]:
    temp.append(dic[age])
feature = np.array(temp).reshape((num_variable,1))
for (N,i) in enumerate(variable[1:]):
    if N == 6:
        feature = np.hstack((feature,np.array(i).reshape((num_variable,1))))
        class_name.append('APR Severity of Illness Code')
    else:
        temp = mlb.fit_transform(zip(i))
        feature = np.hstack((feature,temp))
        class_name += list(mlb.classes_)

order_feature = range(len(feature))
np.random.shuffle(order_feature)
train = feature[order_feature[:1000],:]
test = feature[order_feature[1000:],:]
train_label = label[order_feature[:1000]]
test_label = label[order_feature[1000:]]

clf = RandomForestRegressor(n_estimators = 300,max_features = 0.25)
clf = clf.fit(train, train_label)
p = clf.predict(test)

print('random forest: ' +  str(np.sum(abs(p - test_label))/np.sum(test_label)))
print('')
imp = clf.feature_importances_
for (N,i) in enumerate(imp):
    print(str(class_name[N]) + ': ' + str(i))
    
clf = Ridge(alpha = 36)
clf = clf.fit(train, train_label)
p = clf.predict(test)
print('')
print('redge regression: ' + str(np.sum(abs(p - test_label))/np.sum(test_label)))
print('')
imp = clf.coef_
for (N,i) in enumerate(imp):
    print(str(class_name[N]) + ': ' + str(i))
j_Series = pd.Series(imp,class_name)
```
